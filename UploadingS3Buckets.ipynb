```python
# Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("loadingS3Buckets").getOrCreate()
```

```python
# Read in data from S3 Buckets
from pyspark import SparkFiles
url = "https://first-s3-bucket-data.s3.us-east-2.amazonaws.com/token_uniqueness_score_OK_and_TX.csv"
spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("token_uniqueness_score_OK_and_TX.csv"), sep=",", header=True)

# Show DataFrame
df.show()
```
